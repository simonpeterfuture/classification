```python
def removeOutlier(attribute,data):
    lowerbound=attribute.mean() -3*attribute.std()
    upperbound=attribute.mean() +3*attribute.std()
    data1=data[(attribute>lowerbound) & (attribute<upperbound)]
    data=data1.copy()
    return data
```

```python
data = removeOutlier(data['histogram_mode'],data)
data = removeOutlier(data['histogram_mean'],data)
data = removeOutlier(data['histogram_median'],data)
data = removeOutlier(data['histogram_variance'],data)
data = removeOutlier(data['percentage_of_time_with_abnormal_long_term_variability'],data)
data.info()
```

```python
X=data.drop(['fetal_health'], axis=1)
y=data['fetal_health']
```

```python
col_names=list(X.columns)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_ss=sc.fit_transform(X)
X_ss=pd.DataFrame(X_ss, columns=col_names)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_ss,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier 
```

```python
model=GradientBoostingClassifier(n_estimators=100,learning_rate=0.3,max_depth=6,random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
